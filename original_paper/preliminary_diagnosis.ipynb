### The aggregate U.S. real stock market return ( `ret_t` )

Цитата из статьи:

The aggregate U.S. real stock market return ( t ret )
is obtained by subtracting the CPI inflation rate from the log returns on the CRSP valueweighted market portfolio

Значит берем stock market return и делим на ИПЦ

```python
net_t = pd.read_csv("../original_data/processed/stock_market_return.csv")
net_t['date'] = pd.to_datetime(net_t['date'])
```

```python
cpi = pd.read_csv("../original_data/processed/cpi_us.csv")
cpi['date'] = pd.to_datetime(cpi['date'])
```

```python
net_t = net_t.sort_values('date')
cpi = cpi.sort_values('date')
```

```python
cpi['inflation'] = 100 * np.log(cpi['cpi_us'] / cpi['cpi_us'].shift(1))
```

```python
ret_t = pd.merge(net_t, cpi[['date', 'inflation']], on='date', how='inner')
```

```python
ret_t['ret_t'] = ret_t['stock_market_return'] - ret_t['inflation']
```

```python
ret_t = ret_t[['date', 'ret_t']]
```
